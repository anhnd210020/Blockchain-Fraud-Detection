```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Đường dẫn đến file CSV
file_path = '/home/ducanhhh/Dynamic_Feature-main/ERC-20 Log/Dataset/exp_erc20_trans.csv'

# Đọc dữ liệu
df = pd.read_csv(file_path)
print(f"Dữ liệu có {df.shape[0]:,} dòng và {df.shape[1]} cột.")
df.head()
```

```python
# Thông tin tổng quan
df.info()

# Kiểm tra giá trị thiếu
print("\nSố lượng giá trị thiếu:")
print(df.isnull().sum())

# Mô tả thống kê
df.describe(include='all')
```

```python
# Giả sử cột nhãn là 'label' hoặc 'isp' hoặc 'is_phishing'
if 'label' in df.columns:
    label_col = 'label'
elif 'isp' in df.columns:
    label_col = 'isp'
elif 'is_phishing' in df.columns:
    label_col = 'is_phishing'
else:
    label_col = None

if label_col:
    print(df[label_col].value_counts())
    sns.countplot(data=df, x=label_col)
    plt.title("Phân bố lớp (Label distribution)")
    plt.show()
else:
    print("Không tìm thấy cột nhãn để phân tích phân phối.")
```
